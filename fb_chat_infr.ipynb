```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

from google.colab import drive
drive.mount('/content/drive')

# model path
model_path = "drive/MyDrive/Colab Notebooks/fb_chat2"
```

```python
from transformers import TextGenerationPipeline

# load model
if True:
    from unsloth import FastLanguageModel
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name = model_path,
        max_seq_length = max_seq_length,
        dtype = dtype,
        load_in_4bit = load_in_4bit,
    )
    FastLanguageModel.for_inference(model) # Enable native 2x faster inference
```

```python
# create pipeline
class MyPipeline(TextGenerationPipeline):
    def postprocess(pipeline ,dict):
      # dict_keys(['generated_sequence', 'input_ids', 'prompt_text'])
      tmp = tokenizer.batch_decode([dict['generated_sequence'][0][0]], skip_special_tokens=True)

      # return only 1 message
      ans = tmp[0][len(dict['prompt_text']):]
      chat = ans.strip().split('###')

      res=[]
      answerer = 'Tsu-yun Chen'
      for msg in chat:
        msg = msg.split(':')
        if len(msg) == 2:
          if msg[0].strip() == answerer:
            res.append(msg[1].strip())
          elif len(res)>0:
            break

      return '\n'.join(res)

FastLanguageModel.for_inference(model)
pipe = MyPipeline(
    task="text-generation",
    model = model,
    tokenizer = tokenizer,
    batch_size = 4,
    eos_token_id = model.config.eos_token_id,
    max_new_tokens = 64,
)
```

```python
class conversation:

  def __init__(self, pipe):
    self.log = '<<SYS>>Generate a chat conversation, avoid repeating content<</SYS>>\n'
    self.pipe = pipe

  def response(self, msg):
    msg = f' ### User: {msg}'
    prpt = self.log + msg
    res = pipe(prpt)
    self.log += msg + ' ' + res
    return res

```

```python

talk1 = conversation(pipe)

while True:
  you = input('### User: ')
  print(f'### Tsu-yun Chen: {talk1.response(you)}')
```
